```python
#.... Clean the data.......

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_excel("Customers_Data.xlsx")
print(df.shape)  # 1006 rows, 63 columns

# Remove exact duplicates (3 rows) and conflicting duplicate IDs
df = df.drop_duplicates()
print("After remove exact duplicates:", df.shape)  # 1003 rows
df = df.drop_duplicates(subset=["Customer ID"], keep='first')
print("Unique Customer IDs:", df.shape)  # 1000 rows

# Remove invalid ages (<18 or >100)
df = df[(df['Age'] >= 18) & (df['Age'] <= 100)]
print("After removing unrealistic ages:", df.shape)

# Fix negative financials (likely data errors)
df = df[(df['Estimated Monthly Income (LKR)'] >= 0)]
df = df[(df['Savings Account Balance (LKR)'] >= 0)]
df = df[(df['Last Transaction Date (Days Since)'] >= 0)]
print("After removing negative incomes/balances:", df.shape)

# Fill missing values for numeric product/transaction columns (treat as zero)
num_cols = ['Monthly Credit In - Count','Monthly Credit In - Amount (LKR)',
            'Monthly Debit Out - Count','Monthly Debit Out - Amount (LKR)',
            'Bill Payments - Count','Bill Payments - Amount (LKR)',
            'CC Payments - Count','CC Payments - Amount (LKR)',
            'Inward Remittances - Amount (LKR)',
            'ATM Withdrawals - Count','ATM Withdrawals - Amount (LKR)',
            'POS Transactions - Count','POS Transactions - Amount (LKR)',
            'Interbank Transfers Out (LKR)',
            'Total Transactions per Month','Transaction Days per Month',
            'Mobile App Fund Transfers per Month','Mobile App Bill Payments per Month',
            'Credit Card Utilisation %','FD Tenure (Months)','Loan Outstanding Balance (LKR)',
            'Investment Amount (LKR)','Salary Credit Amount (LKR)']
for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# Standardize categorical flags to consistent values
df['Credit Card'] = df['Credit Card'].astype(str).str.strip().str.upper()
df['Credit Card'] = df['Credit Card'].replace({'Y':'YES','N':'NO','1':'YES','0':'NO', 'TRUE':'YES', 'FALSE':'NO'})
df['Credit Card'] = df['Credit Card'].replace({'YES':'Yes','NO':'No'})
df['Gender'] = df['Gender'].replace({'f':'Female','m':'Male'})
df['Gender'] = df['Gender'].replace({'F':'Female','M':'Male'})
df['Dormancy Flag'] = df['Dormancy Flag'].fillna("No")
df['Churn Risk Flag'] = df['Churn Risk Flag'].fillna("No")
df['Push Notification Opt-in'] = df['Push Notification Opt-in'].fillna("No")
print(df[['Credit Card','Dormancy Flag','Churn Risk Flag','Push Notification Opt-in']].head())


import os

# Folder where you want to save the file
folder_path = "Outputs" # Changed to a relative path suitable for Colab

# File name
file_name = "Cleaned_Customers_Data.xlsx"

# Create the directory if it does not exist
os.makedirs(folder_path, exist_ok=True)

# Create full file path
output_path = os.path.join(folder_path, file_name)

# Export cleaned dataset
df.to_excel(output_path, index=False)

print("Saved successfully:", output_path)
```

```python
#........Feature Engineering & Scoring Model.....


from sklearn.preprocessing import MinMaxScaler

features = ['Estimated Monthly Income (LKR)', 'Avg Balance Last 3 Months (LKR)', 'Savings Account Balance (LKR)',
            'Total Transactions per Month','Transaction Days per Month',
            'Total Products Held',
            'Mobile App Logins per Month',
            'Customer Tenure (Months)']
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[features] = scaler.fit_transform(df[features])

# Compute component scores (0-100 scale)
df_scaled['Financial Score'] = df_scaled[['Estimated Monthly Income (LKR)','Avg Balance Last 3 Months (LKR)','Savings Account Balance (LKR)']].mean(axis=1)*100
df_scaled['Transaction Score'] = df_scaled[['Total Transactions per Month','Transaction Days per Month']].mean(axis=1)*100
df_scaled['Product Score'] = df_scaled['Total Products Held']*100
df_scaled['Digital Score'] = df_scaled['Mobile App Logins per Month']*100
df_scaled['Relationship Score'] = df_scaled['Customer Tenure (Months)']*100

# Weighted total score (0-100)
df_scaled['Customer Score'] = (
    0.35*df_scaled['Financial Score'] +
    0.25*df_scaled['Transaction Score'] +
    0.20*df_scaled['Product Score'] +
    0.10*df_scaled['Digital Score'] +
    0.10*df_scaled['Relationship Score']
)
print(df_scaled[['Financial Score','Customer Score']].describe().round(1))
```

```python
#...7. Customer Tier Classification....

def assign_tier(score):
    if score >= 60: return "Platinum"
    elif score >= 50: return "Gold"
    elif score >= 40: return "Silver"
    elif score >= 30: return "Bronze"
    else: return "Black"

df_scaled['Tier'] = df_scaled['Customer Score'].apply(assign_tier)
print(df_scaled['Tier'].value_counts())




import matplotlib.pyplot as plt

tier_counts = df_scaled['Tier'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    tier_counts,
    labels=tier_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Customer Distribution by Tier")
plt.axis('equal')
plt.show()

# import os

# # Folder where you want to save the file
# folder_path = "Outputs"

# # File name
# file_name = "Customer_Scoring_With_Tiers.xlsx"

# # Create the directory if it does not exist
# os.makedirs(folder_path, exist_ok=True)

# # Create full file path
# output_path = os.path.join(folder_path, file_name)

# # Export customer scoring dataset with all scores and tiers
# df_scaled.to_excel(output_path, index=False)

# print("Saved successfully:", output_path)

#print(df_scaled[['Tier']].describe().round(1))



```

```python
# Create Tier Summary
df['Tier'] = df_scaled['Tier']

tier_summary = (
    df.groupby('Tier')
    .agg({
        'Estimated Monthly Income (LKR)': 'mean',
        'Avg Balance Last 3 Months (LKR)': 'mean',
        'Total Products Held': 'mean',
        'Total Transactions per Month': 'mean',
        'Mobile App Logins per Month': 'mean'
    })
    .round(0)
)

tier_summary.columns = [
    'Avg Income (LKR)',
    'Avg Balance (LKR)',
    'Avg Products',
    'Total Txn/mo',
    'App Logins/mo'
]

tier_order = ['Platinum', 'Gold', 'Silver', 'Bronze', 'Black']
tier_summary = tier_summary.reindex(tier_order)

print(tier_summary)
```

Tier Uplift Strategies (Next-Level Triggers)

```python
# Example: quantifying a specific gap (Bronze vs Silver avg balances)
silver_mean = df_scaled[df_scaled['Tier']=="Silver"]['Avg Balance Last 3 Months (LKR)'].mean()
bronze_mean = df_scaled[df_scaled['Tier']=="Bronze"]['Avg Balance Last 3 Months (LKR)'].mean()
print("Balance gap Bronze->Silver:", silver_mean - bronze_mean)
```

```python
# Calculate absolute gap between tiers
tier_gap = tier_summary.diff(-1).dropna()

# Convert last 2 columns to percentage difference
tier_gap[['Total Txn/mo', 'App Logins/mo']] = (
    tier_gap[['Total Txn/mo', 'App Logins/mo']] /
    tier_summary[['Total Txn/mo', 'App Logins/mo']].iloc[1:].values
) * 100

# Round values
tier_gap = tier_gap.round(1)

print(tier_gap)
```
